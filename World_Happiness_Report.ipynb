```python
# Get numeric columns
numeric_cols = df.drop('Country', axis=1)

# Create a DataFrame of min and max values
min_max_df = pd.DataFrame({
    'Min': numeric_cols.min(),
    'Max': numeric_cols.max()
})

# Combine into 'min–max' string format
min_max_df['Range'] = min_max_df['Min'].astype(str) + ' – ' + min_max_df['Max'].astype(str)

# Display the result
print(min_max_df[['Range']])
```

```python
print(f"The percentage of missing values in:")
for column in df.columns:
    missing_count = df[column].isna().sum()
    missing_percent = (missing_count / len(df)) * 100
    print(f"'{column}': {missing_percent:.2f}%")
```

Trust in government is too high to just drop.
Therefore we may sub with median or mode.

```python
import seaborn as sns
sns.lineplot(df['Trust in Government']);
```

```python
df['Trust in Government'].describe()
```

No extreme outliers based on the min and max and avg. So we can sub with mean.

```python
avg_trust_in_gov = df['Trust in Government'].mean()
df['Trust in Government'].fillna(avg_trust_in_gov, inplace = True)
```

```python
print(f"The percentage of missing values in:")
for column in df.columns:
    missing_count = df[column].isna().sum()
    missing_percent = (missing_count / len(df)) * 100
    print(f"'{column}': {missing_percent:.2f}%")
```

```python
df.isnull().sum()
```

We now see no missing values for "Trust in Government".

```python
import seaborn as sns
sns.lineplot(df['Perceived Corruption']);
```

We see significant outliers, so we will use median.

```python
df.groupby('Country')['Perceived Corruption'].median()
```

We grouped by country so the median is more accurate. This did not get rid of all missing values as some countries are only in the dataset once.


```python
df['Perceived Corruption'] = df['Perceived Corruption'].fillna(
    df.groupby('Country')['Perceived Corruption'].transform('median')
)
```

```python
print(f"The percentage of missing values in:")
for column in df.columns:
    missing_count = df[column].isna().sum()
    missing_percent = (missing_count / len(df)) * 100
    print(f"'{column}': {missing_percent:.2f}%")
```

##3. Which countries report the highest happiness scores over the last five years?

```python
import plotly.express as px

# Filter data to last 5 years
recent_df = df[df['Year'] >= 2017]

# Group by country and calculate average for ALL metrics
avg_metrics = recent_df.groupby('Country').agg({
    'Happiness': 'mean',
    'Wealth': 'mean',
    'Social Support': 'mean',
    'Life Expectancy': 'mean'
}).reset_index()

# Round values to make hover info more readable
for column in ['Happiness', 'Wealth', 'Social Support', 'Life Expectancy']:
    avg_metrics[column] = avg_metrics[column].round(3)

# Sort and keep top 20 for visibility
top_countries = avg_metrics.sort_values('Happiness', ascending=False).head(20)

# Create the bar chart with hover metrics
fig = px.bar(
    top_countries,
    x='Happiness',
    y='Country',
    orientation='h',
    title='Top 20 Countries by Average Happiness Score (2017–2021)',
    labels={'Happiness': 'Avg. Happiness Score', 'Country': 'Country'},
    color='Happiness',
    color_continuous_scale='Viridis',
    hover_data={
        'Wealth': True,
        'Social Support': True,
        'Life Expectancy': True,
        'Happiness': ':.3f'  # Format to 3 decimal places
    },
    hover_name='Country'
)

# Customize hover template for better formatting
fig.update_traces(
    hovertemplate='<b>%{hovertext}</b><br>' +
                  'Happiness: %{x:.3f}<br>' +
                  'Wealth: %{customdata[0]:.3f}<br>' +
                  'Social Support: %{customdata[1]:.3f}<br>' +
                  'Life Expectancy: %{customdata[2]:.3f}<extra></extra>'
)

fig.update_layout(
    yaxis={'categoryorder': 'total ascending'},
    template='plotly_white',
    height=700
)

fig.show()
```

## 4. Are there significant differences in happiness between high and low social support countries?

#Machine Learning

We want to predict the happiness score/level/life ladder using the other factors.


```python
df.columns
```

```python
df["Happiness"].hist()
```

```python
x = df.drop(["Happiness", "Trust in Government"], axis=1)
y = df["Happiness"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
num_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
# Regression Models Cell
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define regression models
models = {
    'LinearRegression': LinearRegression(),  # no randomness
    'KNeighborsRegressor': KNeighborsRegressor(),  # deterministic
    'DecisionTreeRegressor': DecisionTreeRegressor(random_state=42),
    'RandomForestRegressor': RandomForestRegressor(random_state=42),
}

# Train and evaluate each model
for model_name, model in models.items():
    model.fit(X_train[num_cols], Y_train)
    Y_pred = model.predict(X_test[num_cols])
    print(f"\n{model_name}")
    print(f"R² Score: {r2_score(Y_test, Y_pred):.3f}")
    print(f"Mean Squared Error: {mean_squared_error(Y_test, Y_pred):.3f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Define parameter grid for RandomForestRegressor
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['auto', 'sqrt']
}

for model_name, model in models.items():
    if model_name == 'RandomForestRegressor':
        print(f"\nTuning hyperparameters for {model_name}...")
        grid_search = GridSearchCV(estimator=model,
                                   param_grid=param_grid_rf,
                                   cv=5,
                                   scoring='r2',
                                   n_jobs=-1)
        grid_search.fit(X_train[num_cols], Y_train)
        print(f"Best parameters: {grid_search.best_params_}")
        best_model = grid_search.best_estimator_
        Y_pred = best_model.predict(X_test[num_cols])
        print(f"R² Score: {r2_score(Y_test, Y_pred):.3f}")
        print(f"Mean Squared Error: {mean_squared_error(Y_test, Y_pred):.3f}")
    else:
        # Train and evaluate other models without tuning
        model.fit(X_train[num_cols], Y_train)
        Y_pred = model.predict(X_test[num_cols])
        print(f"\n{model_name}")
        print(f"R² Score: {r2_score(Y_test, Y_pred):.3f}")
        print(f"Mean Squared Error: {mean_squared_error(Y_test, Y_pred):.3f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up subplots: 2 rows x 3 columns (adjust as needed)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 10))
axes = axes.flatten()  # Flatten so we can index easily

# Iterate over models and plot in each subplot
for i, (model_name, model) in enumerate(models.items()):
    model.fit(X_train[num_cols], Y_train)
    Y_pred = model.predict(X_test[num_cols])

    ax = axes[i]
    sns.scatterplot(x=Y_test, y=Y_pred, alpha=0.6, ax=ax)
    ax.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--')
    ax.set_xlabel("Actual")
    ax.set_ylabel("Predicted")
    ax.set_title(f"{model_name}")

# Remove unused subplot if models < subplots (5 models, 6 plots)
if len(models) < len(axes):
    for j in range(len(models), len(axes)):
        fig.delaxes(axes[j])

fig.suptitle("Actual vs Predicted Happiness Scores by Model", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to fit suptitle
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

for model_name, model in models.items():
    model.fit(X_train[num_cols], Y_train)
    Y_pred = model.predict(X_test[num_cols])

    # Cross-validation
    cv_scores = cross_val_score(model, X_train[num_cols], Y_train, cv=5, scoring='r2')

    print(f"\n{model_name}")
    print(f"R² Score: {r2_score(Y_test, Y_pred):.3f}")
    print(f"Mean Squared Error: {mean_squared_error(Y_test, Y_pred):.3f}")
    print(f"CV R² Scores: {cv_scores.round(3)}")
    print(f"Average CV R² Score: {cv_scores.mean():.3f}")
```
